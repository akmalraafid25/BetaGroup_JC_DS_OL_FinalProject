# **Problem Statement**

**Business Understanding:**

Bank marketing memiliki peran yang krusial dalam menarik nasabah baru, mempertahankan nasabah, dan mempertahankan image brand di pasar yang kompetitif. Bank marketing muncul mengikuti perkembangan marketing secara umum dan services marketing. Bank marketing adalah proses yang melibatkan bank secara langsung dan proaktif dalam memenuhi kebutuhan dari nasabah sehingga dapat memenuhi kebutuhan dan keinginan bank.

Deposit merupakan hal yang sangat penting dari sudut pandang perbankan. Deposit merupakan dana yang disimpan oleh nasabah di bank dengan jangka waktu tertentu yang bertujuan untuk keamanan dan potensi keuntungan melalui bunga atau pembagian hasil (return). Deposit ini dapat digunakan oleh bank untuk dipinjamkan kepada nasabah lain atau menginvestasikan dana ini untuk mendapatkan keuntungan yang lebih besar bagi bank tersebut. Dana ini juga digunakan oleh bank untuk menutupi pengeluaran operasional dan menjaga kestabilan arus keuangan bagi bank tersebut. Dengan demikian, bank dapat terus berjalan lancar sambil memenuhi kebutuhan nasabah.Melalui bank marketing, bank dapat menarik nasabah untuk menyimpan dana melalui produk deposit yang tersedia. 

Sebuah bank di Portugal menjalankan beberapa kampanye pemasaran melalui panggilan telepon (telemarketing) dengan tujuan utama untuk mempromosikan produk deposito berjangka (term deposit) kepada nasabah dan calon nasabah. Kampanye ini dilakukan dalam berbagai periode, dengan strategi pemasaran langsung yang sangat bergantung pada interaksi personal via telepon.

Dalam konteks ini, telemarketing dilakukan untuk menarik, melibatkan dan mempertahankan nasabah dengan mempromosikan produk dan layanan bank serta membangun hubungan jangka panjang yang kuat sehingga mendukung dan menstabilkan keuangan bank. Pada dataset ini, Hasil dari kampanye ini ketika nasabah setuju untuk melakukan deposit target variabel akan diberi tanda `yes`, sebalikanya `no`

**Problem Statement**

Maka dari itu, perlu adanya efisiensi dari segi biaya dan juga pemahaman terkait calon nasabah yang memiliki probabilitas tinggi untuk melakukan term deposit.

### Checking Missing Values

Karena missing values ditandai dengan nilai `unknown`, maka kita perlu cek kolom mana saja yang memiliki nilai `unknown`

```python
columns_with_value = []
for col in df.columns:
    if 'unknown' in df[col].values:
        columns_with_value.append(col)
print(f"Columns containing the value 'unknown': {columns_with_value}")
```

```python
missing = []
for cols in columns_with_value:
    missing.append([cols, df[df[cols]=='unknown'].shape[0],(df[df[cols]=='unknown'].shape[0]/41188)*100])
missing_values = pd.DataFrame(columns=['Column Name', 'Number of missing', 'Percentage'],
                     data=missing)
missing_values
```

Insights:
- Terdapat 330 missing value (0.8%) pada kolom job
- Terdapat 80 missing value (0.19%) pada kolom marital
- Terdapat 1730 missing value (4.2%) pada kolom education
- Terdapat 8596 missing value (20.8%) pada kolom default
- Terdapat 990 missing value (2.4%) pada kolom housing
- Terdapat 990 missing value (2.4%) pada kolom loan

## **Checking Data Distribution**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.describe().columns
```

```python
# Kolom numerik yang ingin divisualisasikan
num_cols = ['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate',
       'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']

plt.figure(figsize=(10,20), facecolor='white')
sns.set(style='whitegrid', palette='Set1')
plotnumber = 1

for features in num_cols:
    ax=plt.subplot(5,2, plotnumber)
    sns.histplot(x=features, data=df)
    plt.title(features)
    plt.tight_layout()
    plotnumber+=1
```

```python
# Ukuran figure
plt.figure(figsize=(18, 22))

# Loop untuk membuat boxplot dan stripplot
for i, col in enumerate(num_cols, 1):
    plt.subplot(8, 2, i)
    sns.boxplot(x=df[col], color='lightblue', fliersize=5)
    sns.stripplot(x=df[col], color='red', size=1, jitter=0.2, alpha=0.3)
    plt.title(f'Outlier Visualization - {col}', fontsize=12)
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

```python
# Cek distribusi menggunakan Shapiro Wilk Test
from scipy.stats import shapiro

for cols in num_cols:
    stat, p = shapiro(df[cols])
    print('Kolom:', cols)
    print('Statistik:', stat)
    print('p-value:', p)
    
    if p > 0.05:
        print("Data terdistribusi normal (gagal tolak H0)")
    else:
        print("Data tidak terdistribusi normal (tolak H0)")
    print('\n')
```

```python
ages = []
for i in range(len(df['age'].unique())):
    ages.append([df['age'].unique()[i], df[df['age']==df['age'].unique()[i]].shape[0], (df[df['age']==df['age'].unique()[i]].shape[0]/41188)*100])
age = pd.DataFrame(columns=['Age', 'Number of occurences', 'percentage'],
                     data=ages)
pd.set_option('max_colwidth', None)
age[:10].sort_values('percentage', ascending=False)
```

- mayoritas panggilan dilakukan saat nr.employed berada diatas 5100
