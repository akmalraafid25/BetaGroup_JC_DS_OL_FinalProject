Produk deposito berjangka (term deposit) merupakan salah satu sumber pendanaan utama bagi industri perbankan. Menurut data dari European Central Bank (ECB), deposito dengan jangka waktu tetap memberi bank fleksibilitas likuiditas lebih tinggi karena dananya tidak dapat ditarik sewaktu-waktu, menjadikannya lebih stabil dibandingkan tabungan biasa. Hal ini menjadikan peningkatan penjualan deposito sebagai prioritas strategis dalam pengelolaan dana bank (https://data.ecb.europa.eu).

Dataset yang digunakan dalam proyek ini berasal dari kampanye pemasaran langsung sebuah bank di Portugal yang dilakukan melalui telemarketing antara Mei 2008 hingga November 2010, dan telah dianalisis dalam publikasi ilmiah oleh Moro et al. (2014). Dalam periode tersebut, terdapat 41.188 interaksi telepon yang direkam, namun hanya sekitar 3.924 pelanggan (8,8%) yang akhirnya membuka deposito berjangka (https://archive.ics.uci.edu/ml/datasets/Bank+Marketing; Moro et al., 2014).

Sebagai ilustrasi berbasis data: jika bank menghubungi 1.000 calon pelanggan, maka rata-rata hanya sekitar 88 pelanggan yang berhasil dikonversi menjadi pembuka deposito.
Berdasarkan laporan statistik dari Banco de Portugal, rata-rata saldo deposito berjangka individu (retail term deposit) selama periode 2008‚Äì2010 berada pada kisaran ‚Ç¨4.500 hingga ‚Ç¨6.000 per nasabah. Oleh karena itu, angka ‚Ç¨5.000 digunakan dalam proyek ini sebagai estimasi konservatif dan berbasis data (https://www.bportugal.pt/en/page/statistics).

Dengan demikian, peningkatan efektivitas kampanye sebesar hanya 2 poin persentase‚Äîmisalnya dari 8,8% menjadi 10,8%‚Äîdapat menambah sekitar 20 nasabah baru dari setiap 1.000 kontak. Dengan nominal deposito rata-rata ‚Ç¨5.000 per pelanggan, maka potensi tambahan dana terhimpun adalah ‚Ç¨100.000, hanya dari satu batch kampanye.

2. **Predictive Objective: Machine Learning Application**
<br><br>Tujuan dari proyek ini adalah membangun model machine learning untuk membantu bank menentukan apakah seorang pelanggan akan melakukan term deposit, berdasarkan informasi yang tersedia **hingga saat ini**.

    Dataset yang digunakan merepresentasikan hasil dari interaksi telemarketing terakhir terhadap masing-masing pelanggan. Masing-masing baris dianggap mewakili **satu pelanggan unik**, tanpa riwayat kontak terpisah per individu.

    **Model tidak terbatas hanya untuk prediksi kontak pertama**

    Meskipun fitur seperti `campaign` menunjukkan jumlah total kontak yang sudah dilakukan terhadap seorang pelanggan, informasi ini justru merefleksikan **kondisi nyata saat ini** misalnya, "ini adalah kontak ke-2 atau ke-4 untuk pelanggan A." Karena itu:

    - Model **tidak hanya memprediksi pada kontak pertama**
    - Tapi juga memprediksi hasil kontak **saat ini**, terlepas dari sudah ke berapa kali pelanggan dihubungi

    **Justifikasi Pendekatan Ini:**

    - **Tidak ada ID pelanggan** ‚Üí kita tidak bisa membedakan riwayat per orang
    - **Tidak ada baris eksplisit untuk tiap urutan kontak** ‚Üí dataset hanya menyimpan snapshot terakhir per pelanggan
    - Maka, **satu baris = satu pelanggan** adalah asumsi paling aman dan praktis
    - Fitur `campaign`, `previous`, dan `poutcome` tetap bisa digunakan karena nilainya sudah tersedia **sebelum prediksi dilakukan**

    **Tujuan kedua adalah membangun model prediktif berbasis machine learning yang mampu:**

    > Memprediksi apakah pelanggan akan melakukan term deposit pada saat mereka dihubungi, dengan mempertimbangkan informasi historis dan status saat ini tanpa menggunakan data masa depan seperti `duration`.

    Model ini akan digunakan untuk:
    - Membantu tim marketing memprioritaskan pelanggan yang paling potensial
    - Mengurangi biaya kampanye dengan menyaring pelanggan yang kemungkinan besar tidak tertarik

Dengan pendekatan ini, bank diharapkan tidak hanya mendapatkan insight retrospektif, tetapi juga kemampuan untuk mengoptimalkan keputusan kampanye secara proaktif ke depannya.

### Checking Missing Values

Karena missing values ditandai dengan nilai `unknown`, maka kita perlu cek kolom mana saja yang memiliki nilai `unknown`

```python
columns_with_value = []
for col in df.columns:
    if 'unknown' in df[col].values:
        columns_with_value.append(col)
print(f"Columns containing the value 'unknown': {columns_with_value}")
```

```python
missing = []
for cols in columns_with_value:
    missing.append([cols, df[df[cols]=='unknown'].shape[0],(df[df[cols]=='unknown'].shape[0]/41188)*100])
missing_values = pd.DataFrame(columns=['Column Name', 'Number of missing', 'Percentage'],
                     data=missing)
missing_values
```

Dalam dataset kampanye pemasaran bank ini, tidak ditemukan missing value secara teknis dalam bentuk null atau NaN. Namun, terdapat sejumlah variabel kategorikal seperti **job, marital, education, default, housing, dan loan** yang mengandung nilai "unknown", yang pada dasarnya berfungsi sebagai placeholder untuk data yang tidak diketahui atau tidak diberikan oleh nasabah.

## **Conversion Rate in Each Category**

### What is Conversion Rate?

Conversion rate adalah persentase pengunjung yang menyelesaikan tindakan yang diinginkan seperti membeli produk, mengisi formulir, dll. Secara sederhana, conversion rate adalah cara untuk mengetahui seberapa efektif upaya pemasaran Anda dalam mengubah pengunjung menjadi pelanggan atau mendorong mereka melakukan tindakan yang Anda inginkan. Conversion rate membantu perusahaan memahami seberapa efektif kampanye pemasaran yang dilakukan dalam mencapai tujuan yang diinginkan. Dengan mengetahui tingkat konversi, perusahaan dapat mengidentifikasi area yang perlu diperbaiki dan mengoptimalkan strategi pemasaran untuk mendapatkan hasil yang lebih baik. 

Rumus Conversion rate:

*Conversion Rate = (Jumlah Konversi / Jumlah Total Pengunjung) * 100%*


Untuk memperjelas conversion rate berdasarkan umur nasabah, kami melakukan binning terhadap data umur yang akan mengelompokan nasabah berdasarkan 6 kategori yaitu:
- 17-24 tahun
- 25-34 tahun
- 35-44 tahun
- 45-54 tahun
- 55-64 tahun
- 65+ tahun

#### Feature Engineering

```python
# Kelompokkan umur ke dalam rentang
bins = [17, 25, 35, 45, 55, 65, 100]
labels = ['17-24', '25-34', '35-44', '45-54', '55-64', '65+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
```

### Age Range

Umur sering dikaitkan dengan conversion rate dalam marketing karena umur adalah salah satu indikator utama yang memengaruhi perilaku konsumen, preferensi, dan cara pengambilan keputusan. Umur berpengaruh besar terhadap pola pikir, kebutuhan, daya beli, dan perilaku digital konsumen. Maka, mengaitkan umur dengan conversion rate membantu marketer menyesuaikan strategi agar lebih tepat sasaran dan efektif.


```python
# Hitung total dan yang y = yes
conversion_data = df.groupby('age_group')['y'].value_counts().unstack().fillna(0)
conversion_data['conversion_rate'] = conversion_data['yes'] / (conversion_data['yes'] + conversion_data['no'])
# Sort berdasarkan conversion rate
conversion_data_sorted = conversion_data.sort_values(by='conversion_rate', ascending=False)

# Visualisasi
plt.figure(figsize=(8, 6))
bars = sns.barplot(
    x=conversion_data_sorted.index,
    y=conversion_data_sorted['conversion_rate'],
    hue=conversion_data_sorted.index,
    order=conversion_data_sorted.index
)

# Tambahkan label persentase di atas setiap batang
for bar in bars.patches:
    height = bar.get_height()
    bars.annotate(f'{height:.1%}',
                  xy=(bar.get_x() + bar.get_width() / 2, height),
                  xytext=(0, 5),  # offset di atas bar
                  textcoords='offset points',
                  ha='center', va='bottom', fontsize=10)

# Format plot
plt.title('Conversion Rate berdasarkan Kelompok Usia (Diurutkan)')
plt.xlabel('Kelompok Usia')
plt.ylabel('Conversion Rate')
plt.ylim(0, 1)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## **Outlier Analysis**

Outlier (nilai ekstrim) dapat mempengaruhi hasil pelatihan model secara signifikan, terutama saat menggunakan teknik scaling seperti `StandardScaler`. Oleh karena itu, kita melakukan eksplorasi outlier pada setiap fitur numerik.

Strategi umum untuk penanganan outlier:
- **Capping / winsorizing**: membatasi nilai ekstrim ke batas P1/P99
- **Log-transform**: untuk mengurangi efek lonjakan di distribusi miring
- **RobustScaler**: alternatif scaling yang tidak sensitif terhadap outlier

## **Methodology (Modeling/Analysis)**

Untuk membandingkan model secara adil, kita menggunakan pendekatan:

- **Stratified train-test split**:
  - Menjaga proporsi target (`yes` ‚âà 11%)

```python
from sklearn.model_selection import train_test_split

# Split data dengan stratifikasi berdasarkan target
X_train, X_test, y_train, y_test = train_test_split(
    X_capped, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# Cek proporsi target setelah split
print("Train set target distribution:")
print(y_train.value_counts(normalize=True).round(4))

print("\nTest set target distribution:")
print(y_test.value_counts(normalize=True).round(4))
```

Pipeline Modeling dengan Resampler

Karena target kita tidak seimbang (`yes` hanya ~11%), kita perlu menangani imbalance agar model tidak bias ke kelas mayoritas. Untuk itu, kita menggunakan beberapa resampler untuk di uji coba:
- smote
- random over sampler (ROS)
- random under sampler (RUS)
- nearmiss
- smotenc


```python
from imblearn.over_sampling import SMOTE, RandomOverSampler, SMOTENC
from imblearn.under_sampling import RandomUnderSampler, NearMiss
from imblearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import pandas as pd
import numpy as np
import warnings

# ‚úÖ Hilangkan warning XGBoost & FutureWarning
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

# === 1) Pisahkan kolom berdasarkan tipe ===
ordinal_cols = ['education', 'month', 'day_of_week']
nominal_cols = ['job', 'marital', 'default', 'housing', 'loan', 'contact', 'poutcome', 'pdays_grouped']

# === 2) Preprocessor dengan remainder='passthrough' ===
numeric_pipeline = RobustScaler()
ordinal_pipeline = OrdinalEncoder()
nominal_pipeline = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

preprocessor = ColumnTransformer([
    ('num', numeric_pipeline, num_cols),
    ('ordinal', ordinal_pipeline, ordinal_cols),
    ('nominal', nominal_pipeline, nominal_cols)
], remainder='passthrough')

# === 3) Model dan resampler ===
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "KNN": KNeighborsClassifier()
}

resamplers = ["SMOTE", "RandomOverSampler", "RandomUnderSampler", "NearMiss", "SMOTENC"]

# === 4) Define StratifiedKFold dan scoring ===
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)  # ‚úÖ Define skf
scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']

# === 5) Loop modelling yang diperbaiki ===
results = []

for resampler_name in resamplers:
    print(f"\nüîÅ Resampler: {resampler_name}")

    for model_name, model in models.items():
        print(f"   ‚öôÔ∏è Model: {model_name}")

        if resampler_name == "SMOTENC":
            # === Pipeline untuk SMOTENC ===
            # 1. Preprocess data untuk mendapatkan feature names
            X_processed = preprocessor.fit_transform(X_train)
            feature_names = preprocessor.get_feature_names_out()
            
            # 2. Buat mask untuk kolom kategorikal (ordinal + nominal)
            categorical_mask = []
            for i, name in enumerate(feature_names):
                if any(col in name for col in ordinal_cols + nominal_cols):
                    categorical_mask.append(i)
            
            # 3. Pipeline dengan SMOTENC
            pipe_smotenc = Pipeline([
                ('preprocessor', preprocessor),
                ('resampler', SMOTENC(categorical_features=categorical_mask, random_state=42)),
                ('model', model)
            ])
            
            # 4. Cross validation
            cv_scores = cross_validate(pipe_smotenc, X_train, y_train, cv=skf, scoring=scoring, n_jobs=-1)
            
            # 5. Fit dan predict
            pipe_smotenc.fit(X_train, y_train)
            y_pred = pipe_smotenc.predict(X_test)
            y_proba = pipe_smotenc.predict_proba(X_test)[:, 1] if hasattr(pipe_smotenc.named_steps['model'], "predict_proba") else None

            results.append({
                'Resampler': resampler_name,
                'Model': model_name,
                'CV Accuracy Mean': cv_scores['test_accuracy'].mean(),
                'CV F1 Mean': cv_scores['test_f1'].mean(),
                'CV Recall Mean': cv_scores['test_recall'].mean(),
                'CV Precision Mean': cv_scores['test_precision'].mean(),
                'CV ROC AUC Mean': cv_scores['test_roc_auc'].mean(),
                'Test Accuracy': accuracy_score(y_test, y_pred),
                'Test Precision': precision_score(y_test, y_pred),
                'Test Recall': recall_score(y_test, y_pred),
                'Test F1': f1_score(y_test, y_pred),
                'Test ROC AUC': roc_auc_score(y_test, y_proba) if y_proba is not None else None
            })

        else:
            # === Pipeline untuk resampler normal ===
            if resampler_name == "SMOTE":
                resampler = SMOTE(random_state=42)
            elif resampler_name == "RandomOverSampler":
                resampler = RandomOverSampler(random_state=42)
            elif resampler_name == "RandomUnderSampler":
                resampler = RandomUnderSampler(random_state=42)
            elif resampler_name == "NearMiss":
                resampler = NearMiss()
            else:
                raise ValueError("Resampler tidak dikenal")

            pipe = Pipeline([
                ('preprocessor', preprocessor),
                ('resampler', resampler),
                ('model', model)
            ])

            cv_scores = cross_validate(pipe, X_train, y_train, cv=skf, scoring=scoring, n_jobs=-1)

            pipe.fit(X_train, y_train)
            y_pred = pipe.predict(X_test)
            y_proba = pipe.predict_proba(X_test)[:, 1] if hasattr(pipe.named_steps['model'], "predict_proba") else None

            results.append({
                'Resampler': resampler_name,
                'Model': model_name,
                'CV Accuracy Mean': cv_scores['test_accuracy'].mean(),
                'CV F1 Mean': cv_scores['test_f1'].mean(),
                'CV Recall Mean': cv_scores['test_recall'].mean(),
                'CV Precision Mean': cv_scores['test_precision'].mean(),
                'CV ROC AUC Mean': cv_scores['test_roc_auc'].mean(),
                'Test Accuracy': accuracy_score(y_test, y_pred),
                'Test Precision': precision_score(y_test, y_pred),
                'Test Recall': recall_score(y_test, y_pred),
                'Test F1': f1_score(y_test, y_pred),
                'Test ROC AUC': roc_auc_score(y_test, y_proba) if y_proba is not None else None
            })

# === Output ===
results_df = pd.DataFrame(results).sort_values(by='Test F1', ascending=False).reset_index(drop=True)
display(results_df)
```
