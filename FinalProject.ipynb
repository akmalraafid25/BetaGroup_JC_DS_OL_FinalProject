## **Business Understanding**

**Context:**
Bank marketing memiliki peran yang krusial dalam menarik nasabah baru, mempertahankan nasabah, dan mempertahankan image brand di pasar yang kompetitif. Bank marketing muncul mengikuti perkembangan marketing secara umum dan services marketing. Bank marketing adalah proses yang melibatkan bank secara langsung dan proaktif dalam memenuhi kebutuhan dari nasabah sehingga dapat memenuhi kebutuhan dan keinginan bank.

Deposit merupakan hal yang sangat penting dari sudut pandang perbankan. Deposit merupakan dana yang disimpan oleh nasabah di bank dengan jangka waktu tertentu yang bertujuan untuk keamanan dan potensi keuntungan melalui bunga atau pembagian hasil (return). Deposit ini dapat digunakan oleh bank untuk dipinjamkan kepada nasabah lain atau menginvestasikan dana ini untuk mendapatkan keuntungan yang lebih besar bagi bank tersebut. Dana ini juga digunakan oleh bank untuk menutupi pengeluaran operasional dan menjaga kestabilan arus keuangan bagi bank tersebut. Dengan demikian, bank dapat terus berjalan lancar sambil memenuhi kebutuhan nasabah.Melalui bank marketing, bank dapat menarik nasabah untuk menyimpan dana melalui produk deposit yang tersedia. Dalam konteks ini, bank marketing bertujuan untuk menarik, melibatkan dan mempertahankan nasabah dengan mempromosikan produk dan layanan bank serta membangun hubungan jangka panjang yang kuat sehingga mendukung dan menstabilkan keuangan bank.

# **Checking Missing Values, Unique values & Duplicates**

```python
uniques = []
for col in df:
    uniques.append([col, df[col].nunique(), df[col].unique()])
unique_values = pd.DataFrame(columns=['Column Name', 'Number of Unique', 'Unique Sample'],
                     data=uniques)
pd.set_option('max_colwidth', None)
unique_values
```
